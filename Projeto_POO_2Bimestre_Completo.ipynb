## PARTE 4 ‚Äì Polimorfismo e Banco de Dados

**Tarefas:**
- Implementar polimorfismo: sobrescrever um m√©todo na subclasse.

- Implementar inser√ß√£o (INSERT), consulta (SELECT) e exclus√£o (DELETE) de registros no banco de dados.

**C√≥digo:**

```python
import sqlite3

class Aluno_Guilherme:
    def __init__(self, nome_Guilherme, matricula_Guilherme, turma_Guilherme):
        self.__nome_Guilherme = nome_Guilherme
        self.matricula_Guilherme = matricula_Guilherme
        self.turma_Guilherme = turma_Guilherme
        self.faltas_Guilherme = 0

    def get_nome_Guilherme(self):
        return self.__nome_Guilherme

    def set_nome_Guilherme(self, novo_nome_Guilherme):
        self.__nome_Guilherme = novo_nome_Guilherme

    def verificar_situacao_Guilherme(self):
        if self.faltas_Guilherme >= 5:
            return f"{self.__nome_Guilherme} est√° em risco por excesso de faltas!"
        else:
            return f"{self.__nome_Guilherme} est√° com presen√ßa regular."

    def adicionar_falta_Guilherme(self, quantidade):
        self.faltas_Guilherme += quantidade

    def exibir_dados_Guilherme(self):
        print(f"Aluno: {self.__nome_Guilherme}, Matr√≠cula: {self.matricula_Guilherme}, Turma: {self.turma_Guilherme}")

class AlunoMonitor_Guilherme(Aluno_Guilherme):
    def __init__(self, nome_Guilherme, matricula_Guilherme, turma_Guilherme, disciplina_monitoria_Guilherme):
        super().__init__(nome_Guilherme, matricula_Guilherme, turma_Guilherme)
        self.disciplina_monitoria_Guilherme = disciplina_monitoria_Guilherme

    def exibir_dados_Guilherme(self):
        print(f"Monitor: {self.get_nome_Guilherme()}, Matr√≠cula: {self.matricula_Guilherme}, Turma: {self.turma_Guilherme}")
        print(f"Disciplina que monitora: {self.disciplina_monitoria_Guilherme}")

conexao_Guilherme = sqlite3.connect('escola_Guilherme.db')
cursor_Guilherme = conexao_Guilherme.cursor()

cursor_Guilherme.execute('''
CREATE TABLE IF NOT EXISTS alunos_Guilherme (
    id INTEGER PRIMARY KEY AUTOINCREMENT,
    nome TEXT,
    matricula TEXT,
    turma TEXT
)
''')
conexao_Guilherme.commit()

def inserir_aluno_Guilherme(nome, matricula, turma):
    cursor_Guilherme.execute("INSERT INTO alunos_Guilherme (nome, matricula, turma) VALUES (?, ?, ?)",
                             (nome, matricula, turma))
    conexao_Guilherme.commit()
    print(" Aluno inserido com sucesso!")

def listar_alunos_Guilherme():
    cursor_Guilherme.execute("SELECT * FROM alunos_Guilherme")
    alunos = cursor_Guilherme.fetchall()
    print(" Lista de alunos:")
    for aluno in alunos:
        print(aluno)

def deletar_aluno_Guilherme(matricula):
    cursor_Guilherme.execute("DELETE FROM alunos_Guilherme WHERE matricula = ?", (matricula,))
    conexao_Guilherme.commit()
    print(" Aluno exclu√≠do (se existia).")

aluno1_Guilherme = Aluno_Guilherme("Lucas Silva", "202501", "2B")
aluno2_Guilherme = Aluno_Guilherme("Mariana Costa", "202502", "2B")
aluno1_Guilherme.exibir_dados_Guilherme()
aluno2_Guilherme.exibir_dados_Guilherme()

aluno3_Guilherme = Aluno_Guilherme("Carlos Lima", "202503", "2B")
aluno3_Guilherme.set_nome_Guilherme("Carlos Eduardo Lima")
print("Nome do aluno:", aluno3_Guilherme.get_nome_Guilherme())
aluno3_Guilherme.adicionar_falta_Guilherme(3)
print(aluno3_Guilherme.verificar_situacao_Guilherme())
aluno3_Guilherme.adicionar_falta_Guilherme(2)
print(aluno3_Guilherme.verificar_situacao_Guilherme())

aluno_normal = Aluno_Guilherme("Julia Souza", "202504", "2B")
monitor = AlunoMonitor_Guilherme("Fernando Lopes", "202505", "2B", "Matem√°tica")
aluno_normal.exibir_dados_Guilherme()
monitor.exibir_dados_Guilherme()

inserir_aluno_Guilherme("Julia Souza", "202504", "2B")
inserir_aluno_Guilherme("Fernando Lopes", "202505", "2B")
listar_alunos_Guilherme()
deletar_aluno_Guilherme("202504")
listar_alunos_Guilherme()
```

```python
import sqlite3

# Classe principal representando um aluno
class Aluno_Guilherme:
    def __init__(self, nome_Guilherme, matricula_Guilherme, turma_Guilherme):
        # Atributos da classe Aluno
        self.__nome_Guilherme = nome_Guilherme  # Atributo privado
        self.matricula_Guilherme = matricula_Guilherme
        self.turma_Guilherme = turma_Guilherme
        self.faltas_Guilherme = 0

    # M√©todo GET para acessar o nome (encapsulamento)
    def get_nome_Guilherme(self):
        return self.__nome_Guilherme

    # M√©todo SET para alterar o nome (encapsulamento)
    def set_nome_Guilherme(self, novo_nome_Guilherme):
        self.__nome_Guilherme = novo_nome_Guilherme

    # M√©todo para verificar a situa√ß√£o do aluno com base nas faltas
    def verificar_situacao_Guilherme(self):
        if self.faltas_Guilherme >= 5:
            return f"{self.__nome_Guilherme} est√° em risco por excesso de faltas!"
        else:
            return f"{self.__nome_Guilherme} est√° com presen√ßa regular."

    # M√©todo para adicionar faltas ao aluno
    def adicionar_falta_Guilherme(self, quantidade):
        self.faltas_Guilherme += quantidade

    # M√©todo para exibir dados do aluno
    def exibir_dados_Guilherme(self):
        print(f"Aluno: {self.__nome_Guilherme}, Matr√≠cula: {self.matricula_Guilherme}, Turma: {self.turma_Guilherme}")

# Subclasse representando um aluno monitor (herda de Aluno_Guilherme)
class AlunoMonitor_Guilherme(Aluno_Guilherme):
    def __init__(self, nome_Guilherme, matricula_Guilherme, turma_Guilherme, disciplina_monitoria_Guilherme):
        # Usando super() para chamar o construtor da classe base
        super().__init__(nome_Guilherme, matricula_Guilherme, turma_Guilherme)
        self.disciplina_monitoria_Guilherme = disciplina_monitoria_Guilherme

    # Sobrescrevendo o m√©todo exibir_dados_Guilherme para mostrar dados do monitor
    def exibir_dados_Guilherme(self):
        print(f"Monitor: {self.get_nome_Guilherme()}, Matr√≠cula: {self.matricula_Guilherme}, Turma: {self.turma_Guilherme}")
        print(f"Disciplina que monitora: {self.disciplina_monitoria_Guilherme}")

# Conex√£o com o banco de dados SQLite
conexao_Guilherme = sqlite3.connect('escola_Guilherme.db')
cursor_Guilherme = conexao_Guilherme.cursor()

# Criando a tabela de alunos no banco de dados
cursor_Guilherme.execute('''
CREATE TABLE IF NOT EXISTS alunos_Guilherme (
    id INTEGER PRIMARY KEY AUTOINCREMENT,
    nome TEXT,
    matricula TEXT,
    turma TEXT
)
''')
conexao_Guilherme.commit()

# Fun√ß√£o para inserir dados de um aluno no banco de dados
def inserir_aluno_Guilherme(nome, matricula, turma):
    cursor_Guilherme.execute("INSERT INTO alunos_Guilherme (nome, matricula, turma) VALUES (?, ?, ?)",
                             (nome, matricula, turma))
    conexao_Guilherme.commit()
    print(" Aluno inserido com sucesso!")

# Fun√ß√£o para listar todos os alunos do banco de dados
def listar_alunos_Guilherme():
    cursor_Guilherme.execute("SELECT * FROM alunos_Guilherme")
    alunos = cursor_Guilherme.fetchall()
    print(" Lista de alunos:")
    for aluno in alunos:
        print(aluno)

# Fun√ß√£o para deletar um aluno do banco de dados pelo n√∫mero de matr√≠cula
def deletar_aluno_Guilherme(matricula):
    cursor_Guilherme.execute("DELETE FROM alunos_Guilherme WHERE matricula = ?", (matricula,))
    conexao_Guilherme.commit()
    print(" Aluno exclu√≠do (se existia).")

# Testando as funcionalidades
# Criando alunos
aluno1_Guilherme = Aluno_Guilherme("Lucas Silva", "202501", "2B")
aluno2_Guilherme = Aluno_Guilherme("Mariana Costa", "202502", "2B")
aluno1_Guilherme.exibir_dados_Guilherme()
aluno2_Guilherme.exibir_dados_Guilherme()

# Testando encapsulamento e l√≥gica
aluno3_Guilherme = Aluno_Guilherme("Carlos Lima", "202503", "2B")
aluno3_Guilherme.set_nome_Guilherme("Carlos Eduardo Lima")
print("Nome do aluno:", aluno3_Guilherme.get_nome_Guilherme())
aluno3_Guilherme.adicionar_falta_Guilherme(3)
print(aluno3_Guilherme.verificar_situacao_Guilherme())
aluno3_Guilherme.adicionar_falta_Guilherme(2)
print(aluno3_Guilherme.verificar_situacao_Guilherme())

# Testando polimorfismo
aluno_normal = Aluno_Guilherme("Julia Souza", "202504", "2B")
monitor = AlunoMonitor_Guilherme("Fernando Lopes", "202505", "2B", "Matem√°tica")
aluno_normal.exibir_dados_Guilherme()
monitor.exibir_dados_Guilherme()

# Testando banco de dados
inserir_aluno_Guilherme("Julia Souza", "202504", "2B")
inserir_aluno_Guilherme("Fernando Lopes", "202505", "2B")
listar_alunos_Guilherme()
deletar_aluno_Guilherme("202504")
listar_alunos_Guilherme()
```
